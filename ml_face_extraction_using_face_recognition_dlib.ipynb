# Face extraction experiment using face_recognition/dlib library
More info:
* https://github.com/ageitgey/face_recognition/tree/master/examples
* https://github.com/davisking/dlib/tree/master/python_examples

```python
# face_recognition requires cmake command available for dlib installation
import face_recognition
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Read the image for face detection
img_for_recognition = face_recognition.load_image_file('100k-ai-faces-1.jpg')
img = img_for_recognition.copy()

# Detect faces
faces = face_recognition.face_locations(img_for_recognition)

print("Found faces: {}".format(len(faces)))

# Extract faces
extracted_images = []

for face in faces:
    top, right, bottom, left = face
    
    # Draw rectangle around face
    cv2.rectangle(img_for_recognition, (left, top), (right, bottom), (0, 255, 255), 1)
    face_data = img[top:bottom, left:right]
    extracted_images.append(face_data)
    
# Draw original figure with rectangles
plt.figure(figsize=(20,20))
plt.axis("off")
plt.imshow(cv2.cvtColor(img_for_recognition, cv2.COLOR_RGBA2RGB))
plt.show()
```

```python
# Show all extracted images
img_count = len(extracted_images)
print('Faces detected: ', img_count)

print ('Generating images...')
table_size = (int) (math.sqrt(img_count)) + 1

fig = plt.figure(figsize=(20,20))
for i in range(img_count):
    plt.subplot(table_size,table_size,i+1)
    plt.tight_layout()
    plt.imshow(cv2.cvtColor(extracted_images[i], cv2.COLOR_RGBA2RGB), interpolation='none')
    plt.title("Img_{}".format(i + 1))
    plt.xticks([])
    plt.yticks([])

print ('Done.')
```
