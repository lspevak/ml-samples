# Face extraction experiment using OpenCV library

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math

# Load the classifier
# wget --no-check-certificate "https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml"
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Read the input image
img = cv2.imread('100k-ai-faces-1.jpg')
# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

# Extract faces
extracted_images = []

for (x, y, w, h) in faces:
    # Draw rectangle around face
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    face = img[y:y+h, x:x+w]
    extracted_images.append(face)    
    # cv2.imwrite("file{}.jpg".format(i), face)
```

```python
# Draw original figure with rectangles
plt.figure(figsize=(20,20))
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Show all extracted images

img_count = len(extracted_images)
print('Faces detected: ', img_count)

print ('Generating images...')
table_size = (int) (math.sqrt(img_count)) + 1

fig = plt.figure(figsize=(20,20))
for i in range(img_count):
    plt.subplot(table_size,table_size,i+1)
    plt.tight_layout()
    plt.imshow(cv2.cvtColor(extracted_images[i], cv2.COLOR_BGR2RGB), interpolation='none')
    plt.title("Img_{}".format(i + 1))
    plt.xticks([])
    plt.yticks([])

print ('Done.')
```
